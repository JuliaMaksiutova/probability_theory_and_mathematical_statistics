### **1. Даны значения величины заработной платы заемщиков банка (zp) и значения их поведенческого кредитного скоринга (ks): zp = [35, 45, 190, 200, 40, 70, 54, 150, 120, 110], ks = [401, 574, 874, 919, 459, 739, 653, 902, 746, 832]. Найдите ковариацию этих двух величин с помощью элементарных действий, а затем с помощью функции cov из numpy Полученные значения должны быть равны. Найдите коэффициент корреляции Пирсона с помощью ковариации и среднеквадратичных отклонений двух признаков, а затем с использованием функций из библиотек numpy и pandas.**

КОВАРИАЦИЯ

формула ковариации.png

```python
zp = np.array([35, 45, 190, 200, 40, 70, 54, 150, 120, 110])
ks = np.array([401, 574, 874, 919, 459, 739, 653, 902, 746, 832])
```

```python
m_zp = zp.mean()
```

```python
m_ks = ks.mean()
```

```python
M_zpks = (zp * ks).mean()
```

```python
cov_zpks = M_zpks - m_zp * m_ks
```

```python
print(cov_zpks)
```

```python
np.cov(zp, ks, ddof=0)[0, 1]
```

КОРРЕЛЯЦИЯ

```python
std_zp = zp.std()
```

```python
std_ks = ks.std()
```

```python
corr_zpks = cov_zpks / (std_zp * std_ks)
```

```python
print(corr_zpks)
```

```python
np.corrcoef(zp, ks)[0, 1]
```

### **2. Измерены значения IQ выборки студентов, обучающихся в местных технических вузах: 131, 125, 115, 122, 131, 115, 107, 99, 125, 111. Известно, что в генеральной совокупности IQ распределен нормально. Найдите доверительный интервал для математического ожидания с надежностью 0.95.**

Снимок экрана 2023-09-02 174336.png

```python
students_IQ = np.array([131, 125, 115, 122, 131, 115, 107, 99, 125, 111])
```

```python
mean_students_IQ = students_IQ.mean()
```

```python
std_students_IQ = students_IQ.std(ddof=1)
```

```python
mean_std_students_IQ = std_students_IQ / (np.sqrt(len(students_IQ)))
```

```python
t_table_students_IQ = st.t.ppf(0.025, len(students_IQ)-1)
```

```python
print(t_table_students_IQ)
```

```python
print(f"({mean_students_IQ + t_table_students_IQ*(std_students_IQ/len(students_IQ)**0.5)}, {mean_students_IQ - t_table_students_IQ*(std_students_IQ/len(students_IQ)**0.5)})")
```

```python
st.t.interval(confidence=0.95, df=len(students_IQ)-1, loc=np.mean(students_IQ), scale=st.sem(students_IQ))
```
